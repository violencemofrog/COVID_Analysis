```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

# 幂律分布

```python
power_data = pd.read_csv('../data/幂律分布/幂律分布.csv')
```

```python
Florida_color = "#4169E1"
Zhejiang_color = "#FF6347"
Hubei_color = "#00CD66"
```

```python
%matplotlib notebook
date_index = power_data['Date'].values
fig = plt.figure(figsize=(13, 6))

plt.plot(date_index, power_data['Florida power-law index'], color=Florida_color, label='Florida')
plt.plot(date_index, power_data['Zhejiang power-law index'], color=Zhejiang_color, label='Zhejiang')
plt.plot(date_index, power_data['Hubei power-law index'], color=Hubei_color, label='Hubei')

plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(10))
plt.ylabel('Power-law Index')
plt.xlabel('Date')
plt.legend();
```
