```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

# 幂律分布和病例的关系

```python
Florida_case = pd.read_csv('../data/幂律分布和病例关系/FL病例与幂律分布.csv')
Zhejiang_case = pd.read_csv('../data/幂律分布和病例关系/Zhejiang病例与幂律分布.csv')
Hubei_case = pd.read_csv('../data/幂律分布和病例关系/Hubei病例与幂律分布.csv')
```

```python
Florida_color = "#4169E1"
Zhejiang_color = "#FF6347"
Hubei_color = "#00CD66"
Case_color = '#EEEE00'
```

```python
%matplotlib notebook
```

## 佛罗里达

```python
fig_f = plt.figure(figsize=(12, 6))
ax_f = plt.axes()
ax_f.plot(Florida_case['Date'].values, Florida_case['Power-law Index'], color=Florida_color, 
          label='power-law index')

ax_f2 = ax_f.twinx()
ax_f2.plot(Florida_case['Date'].values, Florida_case['New Cases'], color=Case_color,
           label='new cases')

fig_f.gca().xaxis.set_major_locator(ticker.MultipleLocator(15))
ax_f.set_xlabel("Date")
ax_f.set_ylabel("Power-lax Index")
ax_f2.set_ylabel("New Case")
fig_f.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=ax_f.transAxes)
```

## 浙江

```python
fig_z = plt.figure(figsize=(12, 6))
ax_z = plt.axes()
ax_z.plot(Zhejiang_case['Date'].values, Zhejiang_case['Power-law Index'], color=Zhejiang_color, 
          label='power-law index')

ax_z2 = ax_z.twinx()
ax_z2.plot(Zhejiang_case['Date'].values, Zhejiang_case['New Cases'], color=Case_color,
           label='new cases')

fig_z.gca().xaxis.set_major_locator(ticker.MultipleLocator(15))
ax_z.set_xlabel("Date")
ax_z.set_ylabel("Power-lax Index")
ax_z2.set_ylabel("New Case")
fig_z.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=ax_z.transAxes)
```

```python
fig_h = plt.figure(figsize=(12, 6))
ax_h = plt.axes()
ax_h.plot(Hubei_case['Date'].values[:60], Hubei_case['Power-law Index'].values[:60], color=Hubei_color, 
          label='power-law index')

ax_h2 = ax_h.twinx()
ax_h2.plot(Hubei_case['Date'].values[:60], Hubei_case['New Cases'].values[:60], color=Case_color,
           label='new cases')

fig_h.gca().xaxis.set_major_locator(ticker.MultipleLocator(15))
ax_h.set_xlabel("Date")
ax_h.set_ylabel("Power-lax Index")
ax_h2.set_ylabel("New Case")
fig_h.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=ax_h.transAxes)
```
