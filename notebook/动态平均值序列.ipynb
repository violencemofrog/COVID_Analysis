```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 绘制动态平均值序列

```python
florida = pd.read_csv('../data/sequence/FL.csv', header=None)
zhejiang = pd.read_csv('../data/sequence/浙江.csv', header=None)
hubei = pd.read_csv('../data/sequence/湖北.csv', header=None)
```

```python
Florida_color = "#4169E1"
Zhejiang_color = "#FF6347"
Hubei_color = "#00CD66"
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(12, 10))
fig.subplots_adjust(hspace=0.1, wspace=0.1)

ax_f = fig.add_subplot(3, 1, 1)
ax_f.plot(range(florida.size), florida[0].values, color=Florida_color, label='Florida')
ax_f.legend()

ax_z = fig.add_subplot(3, 1, 2)
ax_z.plot(range(zhejiang.size), zhejiang[0].values, color=Zhejiang_color,label='Zhejiang')
ax_z.set_ylabel('Recovery Days');
ax_z.legend()

ax_h = fig.add_subplot(3, 1, 3)
ax_h.plot(range(hubei.size), hubei[0].values, color=Hubei_color,label='Hubei')
ax_h.set_xlabel('Index');
ax_h.legend()
```
